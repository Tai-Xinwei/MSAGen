```python
hi = []
low = []

for name in ['high', 'low']:
    path = f'/blob/shufxi/data/scigpt/solubility_gen/{name}.txt'
    with open(path) as f:
        for line in f:
            line = line.strip()
            if name == 'high':
                hi.append(line)
            else:
                low.append(line)
```

```python
print(len(hi), len(low))
```

```python
hi_unique = list(set(hi))
low_unique = list(set(low))

print(f'high: {len(hi_unique)}')
print(f'low: {len(low_unique)}')
```

```python
# find overlap
overlap = []
for h in hi_unique:
    if h in low_unique:
        overlap.append(h)
print(f'overlap: {len(overlap)}')
```

```python
# remove overlap
hi_unique = [h for h in hi_unique if h not in overlap]
low_unique = [l for l in low_unique if l not in overlap]

print(f'high: {len(hi_unique)}')
print(f'low: {len(low_unique)}')
```
