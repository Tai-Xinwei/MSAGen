```python
import re
train_path = "/msralaphilly2/ml-la/shufxi/data/scigpt/CrystalLLM/train.txt"
valid_folder = "/msralaphilly2/ml-la/yinxia/wu2/shared/SFM/material/structures2/token_ef_sg"
```

```python
from collections import Counter
train_data = []
with open(train_path, "r") as f:
    for line in f:
        line = line.strip().replace("<material>", "").replace("</material>", "")
        line = line.split()
        elements = Counter(line[:-1])
        data = "".join([f"{x}{y}" for x, y in sorted(elements.items())])
        data += f"{line[-1]}"
        train_data.append(data)
```

```python
print(train_data[-2])
```

```python
train_data_dict = {k: True for k in train_data}
```

```python
line="<i>Ca <i>54 <i>162 <i>Yb <i>5 <i>838 <i>P <i>42 <i>O <i>168 <sg161>"
line = line.strip().replace("<i>", "").split()
elements = {}
cnt = 0
last_ele = ""
for i in range(len(line)-1):
    if line[i].isdigit():
        cnt = cnt * 10 + int(line[i])
    else:
        if last_ele != "":
            elements[last_ele] = cnt
        last_ele = line[i]
        cnt = 0
elements[last_ele] = cnt
data = "".join([f"{x}{y}" for x, y in sorted(elements.items())])
data += f"{line[-1]}"
print(data)
```

```python
import glob
import random
valid_data = []
line_cnt = 0
for file in glob.glob(f"{valid_folder}/*"):
    print(file)
    with open(file, "r") as f:
        for line in f:
            elements = {}
            line = line.strip().replace("<i>", "").split()
            last_ele = ""
            for i in range(0, len(line)-1, 2):
                if not line[i].isalpha():
                    elements = {}
                    break
                element = line[i]
                count = line[i+1]
                elements[element] = int(count)
            if len(elements) == 0:
                continue
            sg = line[-1]
            data = "".join([f"{x}{y}" for x, y in sorted(elements.items())])
            data += f"{sg}"
            if data not in train_data_dict:
                line_cnt += 1
                if len(valid_data) < 10000:
                    valid_data.append(line)
                else:
                    j = random.randint(0, line_cnt)
                    if j < 10000:
                        valid_data[j] = line
```

```python
format_valid_data = []
for i in range(len(valid_data)):
    try:
        line = valid_data[i]
        # line = line.replace("<i>", "")
        # line = line.split()
        sg = line[-1]
        seq = ""
        for j in range(0, len(line) - 1, 2):
            element, count = line[j], line[j+1]
            seq += " ".join([element] * int(count)) + " "
        out = f"<material>{seq}{sg}</material>"
        format_valid_data.append(out)
    except Exception as e:
        print(e)
        print(valid_data[i])
```

```python
with open("/msralaphilly2/ml-la/shufxi/data/scigpt/CrystalLLM/valid.txt", "w") as f:
    for line in format_valid_data:
        f.write(line+"\n")
```

```python
import re
# change to new tag
with open("/msralaphilly2/ml-la/shufxi/data/scigpt/CrystalLLM/valid.txt", "r") as f:
    lines = f.readlines()

with open("/msralaphilly2/ml-la/yinxia/wu2/shared/SFM/material/structures2/valid.txt", "w") as f:
    for line in lines:
        line = line.strip()
        line = line.replace("<material>", "").replace("</material>", "")

        fcf_line = f"<fcf>{line}</fcf>"
        f.write(fcf_line+"\n")

        line = line.split()
        sg = line[-1]
        elements = line[:-1]
        counts = {}
        for element in elements:
            if element in counts:
                counts[element] += 1
            else:
                counts[element] = 1
        formula = " ".join(f"{element} {count}" for element, count in counts.items())

        cf1_line = f"<cf1>{formula} {sg}</cf1>"
        f.write(cf1_line+"\n")

        cf2_line = f"<cf2>{formula}</cf2>"
        f.write(cf2_line+"\n")


```
